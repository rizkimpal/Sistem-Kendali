```python
GS = co.tf([1], [1, 1])
print(GS)
```

```python
t = np.linspace(0, 30, 300) 
_, y = co.step_response(GS, t) 
plt.plot(t, y)
plt.xlabel('time (s)')
plt.ylabel('amplitude')
plt.title('step response')
plt.grid()
```

```python
t = np.linspace(0, 30, 300) 
_, y = co.impulse_response(GS, t)
plt.plot(t, y) 
plt.xlabel('time (s)')
plt.ylabel('amplitude')
plt.title('impulse response')
plt.grid()
```

```python
t = np.linspace(0, 30, 300)
_, y = co.forced_response(GS, t) 
plt.plot(t, y) 
plt.xlabel('time (s)')
plt.ylabel('amplitude')
plt.title('ramp response')
plt.grid()
```

```python
def rise_time(a, waktu1, waktu2):
    rt1 = -np.log(waktu1)
    rt2 = -np.log(waktu2)
    return rt1 - rt2
print(rise_time(1, 0.1, 0.9))
```

```python
# Settling time 2%
st = -np.log(0.02)/1
print(st)
```

```python
T = 1
start = 0
stop = 10
increment = 0.1
t = np.arange(start, stop, increment)
y = 1-np.exp(-t/T)
j = 1-np.exp(-st/T)
k = 1-np.exp(-rise_time(1, 0.1, 0.9)/T)
plt.plot(t, y)
plt.title('1st Order Control System')
plt.xlabel('t[s]')
plt.ylabel('y(t)')
plt.grid()
plt.plot(st, j, 'ro') # Titik settling time
plt.annotate('settling time', (st, j))
plt.plot(rise_time(1, 0.1, 0.9), k, 'ro') # Titik rise time
plt.annotate('rise time', (rise_time(1, 0.1, 0.9), k))
```
